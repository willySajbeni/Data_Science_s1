```python
import pandas as pd
import numpy as np
```

```python
# figuring out what data is missing
missing = np.nan
series = pd.Series(['row1','row2',missing,'row4','row5','row6',missing,'row8'])
series
```

```python
# Return True or False if is NaN
series.isnull()
```

```python
# Return Total numbers os NaN
series.isnull().sum()
```

```python
#filling in for missing values
np.random.seed(25)
df = pd.DataFrame(np.random.rand(36).reshape((6,6)))
df
```

```python
# let's fill some points with NaN
# using .loc[] we can fist select the rows, like 2:5 and after comma select a column 0
df.loc[3:5,0] = missing
df.loc[1:4,5] = missing
df
```

```python
# using the .fillna() method, the argument we pass which value we want to fill
df.fillna(df.mean())
```

```python
# here we create a dictionary to fill all the Nan
df.fillna({0:1.5, 5:2.25})
```

```python
df
```

```python
# Filtering out missing values
# here we are dropping all rows with missing values
df.dropna()
```

```python
# here we are dropping all columns with missing values
df.dropna(axis=1)
```

```python
# create a data with duplicate
df = pd.DataFrame({'column 1':[1,1,2,2,3,3,3],
                   'column 2':['a','a','b','b','c','c','c'],
                   'column 3':['A','A','B','B','C','C','C'],})
df
```

```python
# return True or Falso for duplicates row
df.duplicated()
```

```python
# here we go to drop duplicate
df.drop_duplicates()
```

```python
# create a data with duplicate but change de column 3
df = pd.DataFrame({'column 1':[1,1,2,2,3,3,3],
                   'column 2':['a','a','b','b','c','c','c'],
                   'column 3':['A','A','B','B','C','D','C'],})
df
```

```python
# here we are droppping duplicates just from column 3
df.drop_duplicates(['column 3'])
```
