```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.arange(36).reshape((6,6)))
df
```

```python
df2 = pd.DataFrame(np.arange(15).reshape((5,3)))
df2
```

```python
#cancatenating data, columns
pd.concat([df, df2], axis=1)
```

```python
# concatenating data rows
pd.concat([df,df2])
```

```python
# dropping data, rows
df.drop([0,2])
```

```python
# dropping data, column
df.drop([0,2],axis=1)
```

```python
# adding data
# create a series
series = pd.Series(np.arange(6))
series.name = 'added_variable'
series
```

```python
# adding data using .join
variable_add = pd.DataFrame.join(df, series)
variable_add
```

```python
# adding data using .append(), we use ignore_index=False to live the index the same as the were
added_datatable = variable_add.append(variable_add, ignore_index=False)
added_datatable
```

```python
# adding data using .append(), we use ignore_index=True  - continuo index
added_datatable = variable_add.append(variable_add, ignore_index=True)
added_datatable
```

```python
# Sorting data by some column
df_sorted = df.sort_values(by=5,ascending=False)
df_sorted
```

```python
#grouping data by column
data = pd.read_csv('mtcars.csv')
data.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
data.head()
```

```python
#grouping data by column, and take de mean() we group by cyl and than the data became just 3 rows
car_group = data.groupby(data['cyl'])
car_group.mean()
```
