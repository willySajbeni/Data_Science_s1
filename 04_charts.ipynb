```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
```

```python
data = pd.read_csv('mtcars.csv')
data.columns=['car_name', 'mpg', 'cyl', 'disp',	'hp', 'drat', 'wt',	'qsec',	'vs', 'am', 'gear', 'carb']
data.head(2)
```

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
mpg = data.mpg
mpg.plot()
plt.title('MPG')
plt.show()
```

```python
three_variables = data[['cyl','wt','mpg']]
three_variables.plot()
plt.show()
```

```python
plt.bar(x,y)
```

```python
mpg.plot(kind='barh')
```

```python
three_variables.plot(kind='bar')
plt.show()
```
