```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
sb.set_style('whitegrid')
```

```python
data = pd.read_csv('mtcars.csv')
data.columns=['car_name', 'mpg', 'cyl', 'disp',	'hp', 'drat', 'wt',	'qsec',	'vs', 'am', 'gear', 'carb']
data.head(2)
```

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
mpg = data.mpg
mpg.plot()
plt.title('MPG')
plt.show()
```

```python
three_variables = data[['cyl','wt','mpg']]
three_variables.plot()
plt.show()
```

```python
# The object oriented method
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
ax.set_xlim([1,9])
ax.set_ylim([0,5])
ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,3,4,5])
ax.grid()
ax.plot(x,y)
```

```python
fig = plt.figure()
fig, (ax1,ax2,ax3) = plt.subplots(1,3)
ax1.plot(x)
ax2.plot(x,y)
ax3.plot(y)
plt.show()
```

```python
# how change the width on indvidual bar 
wide = [0.2,0.4,0.6,0.9,0.9,0.9,0.6,0.4,0.2]
color = ['salmon']
plt.bar(x,y, width=wide, color=color, align='center')
```

```python
# change more then one color
color_theme = ['darkgray', 'lightsalmon', 'powderblue']
three_variables.plot(color=color_theme)
```

```python
z = [1,2,3,4,0.5]
plt.pie(z)
plt.show()
```

```python
# change the colors and add labes
color_theme = ['#A9A9A9', '#FFA07A', '#B0E0E6', '#FFE4C4', '#BDB76B']
labels = ['bicycle','motorbike','car','van','stroller']
plt.pie(z, colors=color_theme, labels=labels)
plt.show()
```

```python
#Customizing line styles
x1 = range(0,10)
y1 = range(10,0,-1)
plt.plot(x,y,ds='steps', lw=5)
plt.plot(x1,y1, ls='--',lw=10)
plt.show()
```

```python
# Setting plot markers
plt.plot(x,y, marker='1', mew=20)
plt.plot(x1,y1, marker='+',mew=15)
plt.show()
```

```python
# The object oriented method
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(data.car_name, rotation=60, fontsize='medium')
ax.set_title('Miles por gallon of car in mtcars dataset')
ax.set_xlabel('Car names')
ax.set_ylabel('Gallons por miles')
ax.grid()
ax.set_ylim([0,45])
ax.annotate('Toyota Corolla', 
            xy=(19,33.9), 
            xytext=(21,35),
            arrowprops=dict(facecolor='black')
            )
plt.plot()
```

```python
#add legend to your pie
color_theme = ['#A9A9A9', '#FFA07A', '#B0E0E6', '#FFE4C4', '#BDB76B']
labels = ['bicycle','motorbike','car','van','stroller']
plt.pie(z, colors=color_theme)
plt.legend(labels, loc='best')
plt.show()
```

```python
# time series plot
data_sales = pd.read_csv('Superstore-Sales.csv', index_col='Order Date', encoding='cp1252', parse_dates=True)
data_sales.head()
```

```python
data_sales.plot()
plt.show()
```

```python
data_sales_sample = data_sales.sample(n=100, random_state=25, axis=0)
plt.xlabel('Order dta')
plt.ylabel('Order Quantity')
plt.title('Super sales')
data_sales_sample['Order Quantity'].plot()
plt.grid()
plt.show()
```
