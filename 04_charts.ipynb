```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('mtcars.csv')
data.columns=['car_name', 'mpg', 'cyl', 'disp',	'hp', 'drat', 'wt',	'qsec',	'vs', 'am', 'gear', 'carb']
data.head(2)
```

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
mpg = data.mpg
mpg.plot()
plt.title('MPG')
plt.show()
```

```python
three_variables = data[['cyl','wt','mpg']]
three_variables.plot()
plt.show()
```

```python
plt.bar(x,y)
```

```python
mpg.plot(kind='bar')
```

```python
mpg.plot(kind='barh')
```

```python
three_variables.plot(kind='bar')
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
mpg.plot(kind='pie')
```

```python
three_variables.plot(kind='box')
plt.show
```

```python
# how ro save 
three_variables.plot(kind='density')
plt.savefig('density.png')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
ax.set_xlim([1,9])
ax.set_ylim([0,5])
ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,3,4,5])
ax.grid()
ax.plot(x,y)
```

```python
fig = plt.figure()
fig, (ax1,ax2,ax3) = plt.subplots(1,3)
ax1.plot(x)
ax2.plot(x,y)
ax3.plot(y)
plt.show()
```
