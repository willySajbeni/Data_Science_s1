```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('mtcars.csv')
data.columns=['car_name', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
data.index = data.car_name
data.head()
```

```python
data.carb.value_counts()
```

```python
cars_cat = data[['cyl','vs','am','gear','carb']]
cars_cat.head()
```

```python
gear_group = cars_cat.groupby('gear')
gear_group.describe()
```

```python
# transforming variable to categorical data type
data['group'] = pd.Series(data.gear, dtype='category')
```

```python
type(data.group)
```

```python
data.info()
```

```python
data.group.value_counts()
```

```python
# describing categorical data with crosstabs
pd.crosstab(data.am, data.gear)
```
