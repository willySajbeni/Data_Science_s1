```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sb
import scipy as sp
from scipy.stats.stats import pearsonr
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
data = pd.read_csv('mtcars.csv')
data.columns = ['car_name', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
data.head()
```

```python
sb.pairplot(data)
```

```python
X = data[['mpg','hp','qsec','wt']]
sb.pairplot(X)
```

```python
# using scipy to calculate the Pearson correlation coefficient
mpg = data.mpg
hp = data.hp
qsec = data.qsec
wt = data.wt
pearsonr_coefficient, p_value = pearsonr(mpg, hp)
print(f'Pearsonr corelation coefficient {pearsonr_coefficient}')
```

```python
pearsonr_coefficient, p_value = pearsonr(mpg, qsec)
print(f'Pearsonr corelation coefficient {pearsonr_coefficient}')
```

```python
pearsonr_coefficient, p_value = pearsonr(mpg, wt)
print(f'Pearsonr corelation coefficient {pearsonr_coefficient}')
```

```python
# using pandas to caculate the Person correlation coefficient
corr = X.corr()
corr
```

```python
# Using seaborn to visualize the person correlation coefficient
sb.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values)
```
